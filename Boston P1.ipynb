```python
import numpy as np
import pandas as pd
low = pd.DataFrame(pd.read_csv('Blow.csv'))
mid = pd.DataFrame(pd.read_csv('Bmid.csv'))
high = pd.DataFrame(pd.read_csv('Bhigh.csv'))
```

```python
low['par_group'] = 'C'
mid['par_group'] = 'B'
high['par_group'] = 'A'
all_data = low.append(mid)
all_data = all_data.append(high)
```

```python
all_data.sort_values(["Household_Income"], inplace = True)
all_data = all_data.reset_index()
```

```python
all_data = all_data.drop(all_data.index[474:])
```

```python
all_data['income_group'] = 1
all_data.loc[0:158,'income_group'] = 'C'
all_data.loc[158:316,'income_group'] = 'B'
all_data.loc[316:,'income_group'] = 'A'
```

```python
cn=all_data.groupby(by=['par_group','income_group'])
cn.size()
newcn=cn.size()
newcn.reset_index(name='times')
```
